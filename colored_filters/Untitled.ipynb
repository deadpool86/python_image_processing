```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('giraffes.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
sepia_r, sepia_g, sepia_b = (112, 66, 20)
```

```python
sepia = (sepia_r, sepia_g, sepia_b)
```

```python
overlay = np.full(image.shape, sepia, dtype='uint8')
```

```python
overlay = cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(overlay)
```

```python
final_image = cv2.addWeighted(overlay, 0.3, image, 1.0, 0)
```

```python
plt.imshow(final_image)
```

```python
plt.imshow(image)
```
